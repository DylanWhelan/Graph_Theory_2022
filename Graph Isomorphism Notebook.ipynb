# Graph Isomorphism
The Graph Isomorphism problem is the study of whether or not two graphs have an identical structure. This does not mean that the graphs must look the same, rather it means that the graphs must have the same number of edges and vertices and that each point on a graph must have the same connections as it's equivelant on the other graph.

Above is an example of isomorphic graphs, they have the same number of edges and vertices, and most important each point on the graph have the same connections as their equivelants on the other graph. For example Point 1 is connected to 2, 3 and 4, and A is connected to B, C and D.
